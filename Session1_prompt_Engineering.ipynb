```python
import google.generativeai as genai
```

```python
from google.colab import userdata
api_key=userdata.get('Gemini_key_prompt')#code from below of key section
```

```python
genai.configure(api_key=api_key)
```

```python
prompt_template = """
**Context**: Assist in writing short, user-friendly reviews for budget tech products aimed at online buyers on platforms like Amazon, Flipkart, or tech blogs.

**Role**: You are a reliable tech reviewer who explains product pros and cons in a simple, honest way. Your reviews help everyday users understand what to expect before buying.

**Constraints**:
   Word count: 100–150 words
   Tone: Neutral, honest, and helpful
   Include: 1–2 key advantages, 1 major drawback, and a fair conclusion
   Avoid: Over-the-top praise, salesy language, or technical jargon

**Instructions**:
   1. Read the short input summarizing a tech product's price, features, and user opinion.
   2. Identify real-world usage context and highlight the product’s strengths and weakness.
   3. Write a short review explaining who the product is suitable for.
   4. Conclude with a verdict that helps buyers decide confidently.

**Examples**:
Input: “Smartwatch under ₹3000. Accurate tracking, dim screen.”
Output:
- A decent fitness companion at this price. Tracks steps and sleep accurately, but the screen isn’t bright enough outdoors. Great pick for indoor users or fitness starters.

Input: “Bluetooth speaker under ₹1500. Loud sound, no bass.”
Output:
- Delivers clear, loud audio for calls and casual use. Lacks bass depth, but works well for small rooms or podcasts. Good value for light listeners.

Input: “Laptop under ₹40,000. Fast performance, poor webcam.”
Output:
- Fast enough for office tasks, browsing, and students. The webcam isn’t great for video calls, but the overall performance makes it a reliable budget laptop.

Now classify this below
Input: “Tablet under ₹10000. Good display, slow charging.”
Output:
"""
```

```python
models=genai.list_models()
```

```python
list(genai.list_models())#list of models
```

```python
#here we are taking gemini model(gemini-2.0-flash)
model=genai.GenerativeModel('gemini-2.0-flash')#model ready
```

```python
#model ready with the variable name model
#prompt also ready with the name of prompt template
#combined both
```

```python
response=model.generate_content(prompt_template)#model-generate the content by taking my prompt
```
